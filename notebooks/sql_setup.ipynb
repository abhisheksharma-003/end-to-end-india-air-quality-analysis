## Import Required Libraries

The following libraries are used for data handling and database connectivity.

```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
```

## Load Cleaned Dataset

The cleaned dataset generated during preprocessing is loaded into Python before being transferred to MySQL.

```python
df = pd.read_csv("cleaned_aqi.csv")

df.head()
```

## Dataset Overview

Before loading the dataset into MySQL, we inspect its structure and dimensions.

```python
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
print(f"Cities: {df['City'].nunique()}")
```

## Establish MySQL Connection

A SQLAlchemy engine is created to connect Python with the MySQL database.

```python
engine = create_engine(
    "mysql+pymysql://root:YOUR_PASSWORD@localhost/air_quality_project"
)
```

## Import Dataset into MySQL

The cleaned dataset is loaded into the air_quality table.

If the table already exists, it is replaced with the latest version.

```python
df.to_sql(
    "air_quality",
    engine,
    if_exists="replace",
    index=False
)

print("Data successfully loaded into MySQL.")
```
